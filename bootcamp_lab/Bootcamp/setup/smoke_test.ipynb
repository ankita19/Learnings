## Day 2 & Day 3 models (cached, offline)

```python
from transformers import AutoTokenizer, AutoModel, AutoModelForCausalLM
import torch
bert = 'distilbert-base-uncased'
tok = AutoTokenizer.from_pretrained(bert)
enc = AutoModel.from_pretrained(bert, output_attentions=True)
ids = tok('my refund has not arrived', return_tensors='pt')
with torch.no_grad():
    out = enc(**ids)
print('distilbert OK | attention layers:', len(out.attentions))

gtok = AutoTokenizer.from_pretrained('distilgpt2')
gpt = AutoModelForCausalLM.from_pretrained('distilgpt2')
g = gpt.generate(**gtok('Customer:', return_tensors='pt'), max_new_tokens=5,
                 pad_token_id=gtok.eos_token_id)
print('distilgpt2 OK |', gtok.decode(g[0], skip_special_tokens=True))
```
