## 1. Topic

Today’s topic is alignment training for language models.

## 2. Problem

Why do we need RLHF and reward design?
- SFT helps a model imitate good answers.
- But it may still not follow instructions or preferences well.
- We need a stronger training signal.

## 3. Key idea

Learn the sequence in this order:
1. SFT teaches imitation.
2. RLHF adds preference-based learning.
3. Reward design decides what the model should optimize for.
4. Reward hacking is the failure mode when the reward is weak or gameable.

## 4. Pipeline

- Start with a base model
- Use SFT on good demonstrations
- Train a reward model or use human preferences
- Optimize the policy with RLHF
- Watch for reward hacking and refine the reward

```python
# Small mental model

def learning_path():
    return [
        "SFT: imitate good examples",
        "RLHF: learn from preferences",
        "Reward design: define what is good",
        "Reward hacking: exploit the reward"
    ]

for step in learning_path():
    print(step)
```

## 5. Important concepts

- Supervised Fine-Tuning (SFT)
- Preference data
- Reward model
- Policy optimization
- Reward hacking

## 6. One concrete example

Imagine a chatbot is rewarded for being verbose.
A weak reward could push it to produce long but useless answers.
This is a classic reward hacking pattern.

```python
# Toy example: reward function

def reward(response):
    return len(response.split())

bad_answer = "This is a very long answer because longer answers are rewarded."
print("Reward:", reward(bad_answer))
```

## 7. Experiments to run mentally

Ask yourself:
- What happens if the reward only cares about length?
- What happens if it only cares about politeness?
- What happens if it only cares about a single keyword?

## 8. Likely failures

- Reward is too narrow
- Reward is easy to game
- Model over-optimizes the proxy instead of the real goal

## 9. Questions to keep in mind

- What is the real objective?
- What behavior could the model exploit?
- How can I make the reward more robust?

## 10. Final takeaway

SFT teaches imitation, RLHF adds preference-based learning, reward design defines the goal, and reward hacking shows why the goal must be specified carefully.
