```python
# create voting system
# [c1 c1 , c2]

from collections import Counter
def create_voting_system(candidates , top_k):

    if not candidates or len(candidates) < 2 :
        raise ValueError("Atleast two candiadtes are required")
    
    votes_map = Counter(candidates).most_common(top_k)

    return [ candidate for candidate , _ in votes_map ]

```

```python
create_voting_system(['c1', 'c1', 'c2', 'c3'], 1)
```
